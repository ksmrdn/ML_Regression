```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {'x': np.arange(1,101),
        'y': np.hstack((np.zeros(50),np.ones(50)))
}
df = pd.DataFrame(data)
df
```

<hr>

### Logistic Regression
Logit/ Sigmoid function $ \displaystyle \hat{y} = \frac {1} {1+e^{-(mx+c)}} $

```python
from sklearn.linear_model import LogisticRegression
```

```python
logR = LogisticRegression(solver='lbfgs')
logR.fit(df[['x']],df['y'])
m = logR.coef_[0][0]
c = logR.intercept_[0]
```

```python
df['y pred'] = logR.predict(df[['x']])
df['proba'] = 1 / (1 + np.exp(-1*(m*df['x']+c)))
df
```

```python
plt.plot(df['x'], df['y'], 'ro')
plt.plot(df['x'], df['y pred'], 'y-')
plt.plot(df['x'], df['proba'], 'g-')
```
