```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
```

```python
x = np.arange(0,360,10)
y = np.cos(x * np.pi/180) + np.random.normal(0,0.15,len(x))
df = pd.DataFrame({'x': x, 'y': y})
plt.plot(df['x'], df['y'], 'k.')
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
```

```python
lr = LinearRegression()
rd = Ridge()
ls = Lasso()
en = ElasticNet()

lr.fit(df[['x']], df['y'])
rd.fit(df[['x']], df['y'])
ls.fit(df[['x']], df['y'])
en.fit(df[['x']], df['y'])
```

```python
df['y LR'] = lr.predict(df[['x']])
df['y RD'] = rd.predict(df[['x']])
df['y LS'] = ls.predict(df[['x']])
df['y EN'] = en.predict(df[['x']])
```

```python
plt.plot(df['x'], df['y'], 'k.')
plt.plot(df['x'], df['y LR'], 'r-')
plt.plot(df['x'], df['y RD'], 'b-')
plt.plot(df['x'], df['y LS'], 'g-')
plt.plot(df['x'], df['y EN'], 'y-')
plt.legend(['Data', 'Linear', r'Ridge $\alpha=1$', r'Lasso $\alpha=1$', r'ElasticNet $\alpha=1$' ])
plt.grid(True)
```
